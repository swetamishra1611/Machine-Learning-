```python
import pandas as pd
```

```python
sr2=pd.Series([5,6,7,8,9,10])
sr2
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                 'Name':['A','B','C','D'],
                 'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],
                 'Name':['E','F','G','H'],
                 'Class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1],axis=1)
```

```python
pd.concat([df2,df1],axis=0,ignore_index=True)
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                 'Name':['A','B','C','D'],
                 'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[3,4],
                 'Name':['C','D'],
                 'Class':[7,8]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join='inner')
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                 'Name':['A','B','C','D'],
                 'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],
                 'Name':['E','F','G','H'],
                 'Class':[9,10,11,12]})
df2
```

```python
pd.concat([df1,df2],keys=['df1','df2'])
```

```python
pd.concat([df1,df2],keys=[' First df','Second df2'])
```

```python
pd.concat([df1,df2],axis=1,keys=[' First df','Second df2'])
```
