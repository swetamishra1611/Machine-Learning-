```python
import seaborn as sns 
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
tips_df= sns.load_dataset("tips")
tips_df
```

```python
tips_df.total_bill.sort_values()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"],bins=bins)
plt.xticks(bins)
plt.title("Histogram of Total Bill")
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"],hist_kws = {'color': '#DC143C','edgecolor':'#aaff00','linewidth':5,
                                              'alpha':0.9})
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"],hist_kws = {'color': '#DC143C','edgecolor':'#aaff00','linewidth':5,
                                              'alpha':0.9},
             kde_kws = {'color':'#DC143C','linewidth': 5,'linestyle':'--','alpha':0.9},
             rug= True,
             rug_kws={'color':'#0426d0','edgecolor':'#00dbff','linewidth':5,'linestyle':'--',
                     'alpha':0.9}
            )
```
