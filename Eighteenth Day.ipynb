### How to show image using matplotlib imshow function?

```python
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
```

```python
img = mpimg.imread("all plot.png")
img
```

```python
type(img)
```

```python
img.shape
```

```python
img.ndim
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
single_channel = img[:,:,2]
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(single_channel,cmap="hot")
plt.colorbar()
plt.show()
```
