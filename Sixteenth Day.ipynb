### Pie chart

```python
import matplotlib.pyplot as plt
```

```python
classes = ["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_students = [ 45,15,35,25,30]
plt.pie([1])
plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_students = [ 45,15,35,25,30]
plt.pie(class1_students,labels =classes )
plt.show()
```

```python
classes = ["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_students = [ 45,15,35,25,30]
explode= [ 0.1,0,0.23,0,0]
colors = ["c","r",'m','g','b']
textprops = {"fontsize": 15}
plt.pie(class1_students,labels =classes ,explode = explode,colors = colors,
       autopct = "%0.2f%%",shadow = True,radius= 1.2,startangle = 270,textprops = textprops)
plt.show()
```

```python
plt.figure(figsize = (6,8))
wedgeprops = { "linewidth":4,"width":2,"edgecolor":"k"}
plt.pie(class1_students,labels =classes ,explode = explode,colors = colors,
       autopct = "%0.2f%%",pctdistance= 0.6,shadow = True,radius= 1.2,labeldistance= 1.2,startangle = 270,
        textprops = textprops,counterclock=True,wedgeprops = wedgeprops,center = (2,3),
       frame = True,rotatelabels = True)
plt.show()
```

```python
explode= [ 0.1,0,0.23,0,0]
colors = ["c","r",'m','g','b']
textprops = {"fontsize": 15}
plt.pie(class1_students,labels =classes ,explode = explode,colors = colors,
       autopct = "%0.2f%%",shadow = True,radius= 2.7,startangle = 270,textprops = textprops)
plt.legend()
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,4))
colors=['r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w']
labels= np.ones(20)
plt.pie([1],colors="k",radius=2.05)
plt.pie(labels,colors=colors,radius=2.0)
plt.pie([1],colors="g",radius=1.8)
plt.pie([1],colors="m",radius=1.6)
plt.pie([1],colors="y",radius=1.3)
plt.pie([1],colors="b",radius=1.1)
plt.pie(labels,colors=colors,radius=0.7)
plt.pie([1],colors="c",radius=0.2)
plt.show()
```
