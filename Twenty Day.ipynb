### Lineplot using Seaborn

```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
days=[1,2,3,4,5,6]
temperature=[36.2,34,43,42.3,43,44]
temp_df = pd.DataFrame ({"days":days,"temperature": temperature})
sns.lineplot(x="days",y="temperature",data=temp_df)
plt.show()
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.lineplot(x="size",y="total_bill",data=tips_df,hue="sex")
plt.show()
```

```python
sns.lineplot(x="size",y="total_bill",data=tips_df,hue="sex",style="sex",
            palette="hot",dashes=False,markers=["o","<"],legend=True)
plt.title("Line Plot",fontsize=20)
plt.xlabel("Size",fontsize=15)
plt.ylabel("Total Bill",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set(style='darkgrid')
sns.lineplot(x="size",y="total_bill",data=tips_df,hue="sex",style="sex",
            palette="hot",dashes=False,markers=["o","<"],legend=True)
plt.title("Line Plot",fontsize=20)
plt.xlabel("Size",fontsize=15)
plt.ylabel("Total Bill",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set(style='darkgrid')
sns.lineplot(x="size",y="total_bill",data=tips_df,hue="day",style="day",
            palette="hot",dashes=False,markers=["o","<",">","^"],legend="brief")
plt.title("Line Plot",fontsize=20)
plt.xlabel("Size",fontsize=15)
plt.ylabel("Total Bill",fontsize=15)
plt.show()
```

### Histogram/Distplot using Seaborn

```python
sns.distplot(tips_df["size"])
```

```python
sns.distplot(tips_df["tip"])
```

```python
sns.distplot(tips_df["total_bill"])
```

```python
sns.distplot(tips_df["total_bill"],bins=10)
```

```python
sns.distplot(tips_df["total_bill"],bins=100)
```

```python
sns.distplot(tips_df["total_bill"],hist=False)
```

```python
sns.distplot(tips_df["total_bill"],kde=False)
```

```python
sns.distplot(tips_df["total_bill"],rug=True)
```

```python
sns.distplot(tips_df["total_bill"],fit=norm,kde=False)
```

```python
sns.distplot(tips_df["total_bill"],color="r")
```

```python
sns.distplot(tips_df["total_bill"],vertical=True)
```

```python
sns.distplot(tips_df["total_bill"],norm_hist=True)
```

```python
sns.distplot(tips_df["total_bill"],axlabel="Total Bill")
```

```python
sns.distplot(tips_df["total_bill"],label="Total Bill")
plt.title("Histogram of Total bill")
plt.legend()
```
