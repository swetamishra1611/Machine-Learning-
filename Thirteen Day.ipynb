### Line plot

```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5,6,7]
temperature = [36,43.5,42,32,27,45.1,32.7]
plt.plot(days,temperature)
plt.axis([0,10,0,50])
plt.title("Delhi Temperature")
plt.xlabel("days")
plt.ylabel("temperature")
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
temperature = [36,43.5,42,32,27,45.1,32.7]
plt.plot(days,temperature,color="g",marker="o",linestyle="--",linewidth=3
,markersize=10)
plt.axis([0,10,0,50])
plt.title("Delhi Temperature")
plt.xlabel("days")
plt.ylabel("temperature")
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
temperature = [36,43.5,42,32,27,45.1,32.7]
plt.plot(days,temperature,"ro--",linewidth=3
,markersize=10)
plt.axis([0,10,0,50])
plt.title("Delhi Temperature")
plt.xlabel("days")
plt.ylabel("temperature")
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
temperature = [36,43.5,42,32,27,45.1,32.7]
plt.plot(days,temperature,"ro--",linewidth=3
,markersize=10)
plt.axis([0,10,0,50])
plt.title("Delhi Temperature",fontsize=18)
plt.xlabel("days",fontsize=12)
plt.ylabel("temperature",fontsize=12)
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
temperature = [36,43.5,42,32,27,45.1,32.7]
plt.plot(days,temperature,"go--",linewidth=3
,markersize=10)
plt.axis([0,10,0,50])
plt.title("Delhi Temperature",fontsize=18)
plt.xlabel("days",fontsize=12)
plt.ylabel("temperature",fontsize=12)
plt.legend(["Tem line"],loc=4)
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
temperature = [36,43.5,42,32,27,45.1,32.7]
plt.plot(days,temperature,"go--",linewidth=3
,markersize=10,label = "Temp line")
plt.axis([0,10,0,50])
plt.title("Delhi Temperature",fontsize=18)
plt.xlabel("days",fontsize=12)
plt.ylabel("temperature",fontsize=12)
plt.legend(loc=4)
plt.show()
```

```python
from matplotlib import style
days = [1,2,3,4,5,6,7]
temperature = [36,43.5,42,32,27,45.1,32.7]
style.use("ggplot")
plt.plot(days,temperature,"mo--",linewidth=3
,markersize=10,label = "Temp line")
plt.axis([0,10,0,50])
plt.title("Delhi Temperature",fontsize=18)
plt.xlabel("days",fontsize=12)
plt.ylabel("temperature",fontsize=12)
plt.legend(loc=4)
plt.show()
```

```python
from matplotlib import style
days = [1,2,3,4,5,6,7]
temperature = [36,43.5,42,32,27,45.1,32.7]
style.use("ggplot")
plt.plot(days,temperature,"mo--",linewidth=3
,markersize=10,label = "Temp line")
plt.axis([0,10,0,50])
plt.title("Delhi Temperature",fontsize=18)
plt.xlabel("days",fontsize=12)
plt.ylabel("temperature",fontsize=12)
plt.legend(loc=4)
plt.grid( color='r',linestyle='-',linewidth=2)
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
delhi_tem = [36,43.5,42,32,27,45.1,32.7]
mumbai_tem = [32,43,48,21,37,26.8,34.9]
style.use("ggplot")
plt.plot(days,delhi_tem,"go--",linewidth=3
,markersize=10,label = "Delhi tem")
plt.plot(days,mumbai_tem,"ro--",linewidth=3
,markersize=10,label = "Mumbai tem")
plt.axis([0,10,0,50])
plt.title("Delhi Temperature",fontsize=18)
plt.xlabel("days",fontsize=12)
plt.ylabel("temperature",fontsize=12)
plt.legend(loc=4)
plt.grid( color='w',linestyle='-',linewidth=2)
plt.show()
```
